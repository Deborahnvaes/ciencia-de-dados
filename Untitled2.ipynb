```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(50)
y = np.random.rand(50) #50 amostras aleatorias

plt.bar(x, y, color='pink')
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.title("Gr√°fico de barras criado com o Numpy")
plt.show()
```

Stacks Plots

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
```

```python
plt.stackplot(dias,dormir, comer, trabalhar,
              labels=['dias','dormir','comer','trabalhar'],
              colors = ['m','c','b','y'])
plt.legend(loc='upper left')
plt.show() #CORES N√ÉO RECOMEND√ÅVEIS!!!!!!!!!
```
