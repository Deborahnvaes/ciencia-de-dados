```python
#Qual a distribuição de idade dos participantes da pesquisa
#variavel idade

idade = dados.Age #dados['Age']

#histograma

plt.hist(idade, label = "Idades", color='lightseagreen')
plt.title("Distribuição por Idade")
plt.xlabel("Idades")
plt.ylabel("Quantidade")
plt.legend()
plt.grid(True)
plt.show()
```
