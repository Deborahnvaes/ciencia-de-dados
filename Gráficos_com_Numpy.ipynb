**Estudo com  gráficos Numpy**

Obs: Algumas cores não recomendáveis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(50)
y = np.random.rand(50) #50 amostras aleatorias

plt.bar(x, y, color='pink')
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.title("Gráfico de barras criado com o Numpy")
plt.show()
```

```python
x= [1,2,3,4,5,6,7,8]
y= [5,2,6,4,7,6,3,2]

'''
'o' = bolinha
'v' = triangulo
'x' = x
'''
plt.scatter(x, y, label = 'Pontos', color='r', marker = 'x', s = 80)
plt.legend()
plt.show()
```

Stacks Plots

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
```

```python
plt.stackplot(dias,dormir, comer, trabalhar,
              labels=['dias','dormir','comer','trabalhar'],
              colors = ['m','c','k','y'])
plt.legend(loc='upper left')
plt.show() #CORES NÃO RECOMENDÁVEIS!!!!!!!!!
```

Pie Chart ou gráfico de pizza

```python
porcentagens = [70,20,20,130]
cidades = ['São Paulo', 'Recife', 'Fortaleza', 'Manaus']
colunas = ['c', 'm','r','k']
```

```python
plt.pie(porcentagens, labels = cidades, colors = colunas,
        startangle=90, shadow = True, explode = (0,0.15,0,0))
plt.show()
```
